```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

participant_data = pd.read_csv("Demographics Information - Human Participants.csv")
print(participant_data.shape)
participant_data.head()
```

```python
plt.show(sns.displot(x='Present', data=participant_data))

plt.show(sns.displot(x='Present.1', data=participant_data))

# plt.show(sns.displot(x="Present", y="Present.2", data=participant_data, cbar=True))
```

```python
participant_data.groupby("Present").count() # counts for race reporting
```

```python
participant_data.groupby("Present.1").count() # ethnicity reporting
```

```python
participant_data.groupby("Present.2").count() # gender reporting
```

```python
participant_data.groupby("Present.3").count() # age reporting
```

```python
race = [165, 10]
ethnicity = [169, 6]
gender = [54, 121]
age = [50, 125]
x = [0, 1]

x_axis = np.arange(len(x))

plt.bar(x_axis - 0.2, race, 0.1, label = 'Race', color = 'darkorchid')
plt.bar(x_axis - 0.1, ethnicity, 0.1, label = 'Ethnicity', color = 'thistle')
plt.bar(x_axis, gender, 0.1, label = 'Sex', color = 'palevioletred')
plt.bar(x_axis + 0.1, age, 0.1, label = 'Age', color = 'orchid')

plt.xticks(x_axis, x_axis)
plt.xlabel("Variable Present/Absent")
plt.grid(axis='y')
plt.ylabel("Count")
plt.legend()
plt.show()
```
