```python
raceAgeInd=pd.crosstab(participant_data["Present"],participant_data["Present.3"])
print('Race Age table :\n',raceAgeInd)
```

### Chi Squared Test

Run chi2 test on race and age with null hypothesis that the two variables are independent and alternate hypothesis that the two are not. 
alpha = .05

```python
import scipy.stats
from scipy.stats import chi2, chi2_contingency

chi2_stat, p, dof, expected = scipy.stats.chi2_contingency(raceAgeInd)
 
print(f"chi2 statistic:     {chi2_stat:.5g}")
print(f"p-value:            {p:.5g}")
print(f"degrees of freedom: {dof}")
print("expected frequencies:\n",expected)
```

### 95% Confidence Intervals

```python
# Race 95% confidence interval
rdata = participant_data['Present'].to_numpy()
ci_race = stats.norm.interval(alpha = .95, loc = np.mean(rdata), scale = stats.sem(rdata))
print(ci_race)
```

```python
# Ethnicity 95% confidence interval
edata = participant_data['Present.1'].to_numpy()
ci_ethn = stats.norm.interval(alpha = .95, loc = np.mean(edata), scale = stats.sem(edata))
print(ci_ethn)
```

```python
# Gender 95% confidence interval
gdata = participant_data['Present.2'].to_numpy()
ci_gender = stats.norm.interval(alpha = .95, loc = np.mean(gdata), scale = stats.sem(gdata))
print(ci_gender)
```

```python
# Age 95% confidence interval
adata = participant_data['Present.3'].to_numpy()
ci_age = stats.norm.interval(alpha = .95, loc = np.mean(adata), scale = stats.sem(adata))
print(ci_age)
```

Because the 95% Confidence Intervals for Race/Ethnicity do not overlap with those for Age/Gender, there is significant statistical evidence that age and gender variables are reported at a higher frequency than race and ethnicity are.

### Hypotheses Testing

```python
print("Race/Ethnicity:")
print(stats.ttest_ind(rdata, edata))
```

We consider the null hypothesis that the reporting for race equal to the reporting of ethnicity. The alternative hypothesis is that the reporting of race and ethnicity are not equal. We can conduct a two-sided t-test with an &alpha; = .05.

The 2-sided p-value is about .307, which is not statistically significant at the .05 level, thus we cannot reject the null hypothesis. Race and ethnicity are reported at equal freqnecies.


```python
print("Race/Gender:")
print(stats.ttest_ind(rdata, gdata))
```

We consider the null hypothesis that the reporting for race equal to the reporting of gender. The alternative hypothesis is that the reporting of race and gender are not equal. We can conduct a two-sided t-test with an &alpha; = .05.

The 2-sided p-value is about .2.54e-44, which is statistically significant at the .05 level, thus we reject the null hypothesis; race and gender are not reported at equal frequencies.

```python
print("Race/Age:")
print(stats.ttest_ind(rdata, adata))
```

We consider the null hypothesis that the reporting for race equal to the reporting of age. The alternative hypothesis is that the reporting of race and age are not equal. We can conduct a two-sided t-test with an &alpha; = .05.

The 2-sided p-value is about 6.98e-48, which is statistically significant at the .05 level, thus we reject the null hypothesis; race and age are not reported at equal frequencies.

```python
print("Ethnicity/Gender:")
print(stats.ttest_ind(edata, gdata))
```

We consider the null hypothesis that the reporting for ethnicity equal to the reporting of gender. The alternative hypothesis is that the reporting of ethnicity and gender are not equal. We can conduct a two-sided t-test with an &alpha; = .05.

The 2-sided p-value is about .1.78e-49, which is statistically significant at the .05 level, thus we reject the null hypothesis; ethnicity and gender are not reported at equal frequencies.

```python
print("Ethnicity/Age:")
print(stats.ttest_ind(edata, adata))
```

We consider the null hypothesis that the reporting for ethnicity equal to the reporting of age. The alternative hypothesis is that the reporting of ethnicity and age are not equal. We can conduct a two-sided t-test with an &alpha; = .05.

The 2-sided p-value is about 2.31e-53, which is statistically significant at the .05 level, thus we reject the null hypothesis; ethnicity and age are not reported at equal frequencies.

```python
print("Age/Gender:")
print(stats.ttest_ind(adata, gdata))
```

We consider the null hypothesis that the reporting for age equal to the reporting of gender. The alternative hypothesis is that the reporting of age and gender are not equal. We can conduct a two-sided t-test with an &alpha; = .05.

The 2-sided p-value is about .641, which is not statistically significant at the .05 level, thus we cannot reject the null hypothesis. Age and gender are reported at equal frequencies.


If we want to adjust the null hypotheses and alternate hypotheses to be one-tailed (i.e. reported >, reported <=), we can keep these results and divide the calculated p-value from each test by 2 to obtain the 1-tailed p-value. In each case above, the final result is not changed, we find that the tests in which we rejected the two-tailed null hypothesis we will also reject the one-tailed hypothesis, and vice versa.
